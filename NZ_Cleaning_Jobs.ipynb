```python
import pandas as pd
import numpy as np
import re
```

```python
NZ_RawJobFile = pd.read_excel('NZ_Admin_JOBS.xlsx',header=None,skiprows=1)
print (type(NZ_RawJobFile))
print (NZ_RawJobFile.dtypes)
print (NZ_RawJobFile.isna().any())
print(NZ_RawJobFile.shape)
```

```python
NZ_RawJobFile.drop_duplicates(inplace=True)
print(NZ_RawJobFile.shape)
```

```python
NZ_RawJobFile.iloc[:,0] = NZ_RawJobFile.iloc[:,0].str.title()
```

```python
NZ_RawJobFile[NZ_RawJobFile.iloc[:,0].str.contains('Part Time|Part-Time|Fixed Term|Fixed-Term')==True]
```

```python
tables = ('Part Time','Part-Time','Fixed-Term','Fixed Term','Contract','Permanent','9 Month Fixed Term Contract')

t_series = pd.Series(index=NZ_RawJobFile.index)

for a,row in enumerate(NZ_RawJobFile[0]):
    if row.startswith(tables):
       # NZ_RawJobFile.insert(NZ_RawJobFile.iloc[a][1],'new',row[:9])
        if row[:9] in tables[:2]:
            row = row.replace(row[:10],'') #
            t_series[a]='Part Time'
        elif row[:10] in tables[2:4]:
            row = row.replace(row[:11],'')
            t_series[a]='Fixed Term'
        elif row[:9] == 'Contracts':
            row = row.replace(row[:10],'')
            t_series[a]='Contract'
        elif row[:11] =='Contracting':
            row = row.replace(row[:12],'')
            t_series[a]='Contract'
        elif row[:8] =='Contract':
            row = row.replace(row[:9],'')
            t_series[a]='Contract'    
        elif row[:7] == '9 Month':
            row = row.replace(row[:30],'')
            t_series[a]='Fixed Term'    
        else:
            row = row.replace(row[:23],'')
            t_series[a]='Permanent & Fixed Term'
    else:
        t_series[a]='None'
    NZ_RawJobFile[0][a]= row
    
# print(t_series)
# print(NZ_RawJobFile)



  
```

```python

strinfo = re.compile(r'Co-O')
for a,row in enumerate(NZ_RawJobFile[0]):
    d = strinfo.sub('Coo', row)
    NZ_RawJobFile[0][a]= d

    
print(NZ_RawJobFile.iloc[284][0])
    
```
