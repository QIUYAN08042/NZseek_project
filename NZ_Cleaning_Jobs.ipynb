```python
tables = ('Part Time','Part-Time','Fixed-Term','Fixed Term','Contract','Permanent','9 Month Fixed Term Contract')

t_series = pd.Series(index=NZ_RawJobFile.index)

for a,row in enumerate(NZ_RawJobFile[0]):
    if row.startswith(tables):
       # NZ_RawJobFile.insert(NZ_RawJobFile.iloc[a][1],'new',row[:9])
        if row[:9] in tables[:2]:
            row = row.replace(row[:10],'') #去掉空格
            t_series[a]='Part Time'
        elif row[:10] in tables[2:4]:
            row = row.replace(row[:11],'')
            t_series[a]='Fixed Term'
        elif row[:9] == 'Contracts':
            row = row.replace(row[:10],'')
            t_series[a]='Contract'
        elif row[:11] =='Contracting':
            row = row.replace(row[:12],'')
            t_series[a]='Contract'
        elif row[:8] =='Contract':
            row = row.replace(row[:9],'')
            t_series[a]='Contract'    
        elif row[:7] == '9 Month':
            row = row.replace(row[:30],'')
            t_series[a]='Fixed Term'    
        else:
            row = row.replace(row[:23],'')
            t_series[a]='Permanent & Fixed Term'
    else:
        t_series[a]='None'
    NZ_RawJobFile[0][a]= row
    
# print(t_series)
# print(NZ_RawJobFile)



  
```

```python
def time_transfer(x):
    if 'd' in x:
        return -int(x[:-1])
    elif 'm' in x:
        return -(int(x[:-1])*30)
    elif x =='Unknown':
        return 'Unknown'
    else:
        return 0
       
```

```python
NZ_RawJobFile['Post_Time']=NZ_RawJobFile['Post_Time'].apply(time_transfer)
```

```python
def delduplicated(x):
    sentence=x.strip()
    index=(sentence+sentence).find(sentence,1)
    if index!=-1:
        return sentence[:index]
    else:
        return 'Unknown'
```
